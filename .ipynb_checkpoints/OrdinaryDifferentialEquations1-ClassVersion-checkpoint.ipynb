# Ordinary Differential Equations

## Introduction
What is an ordinary differential equation?

We've had an example already when trying to define the exponential function. In fact, we motivated the exponential function from the basic example of an equation that describes population growth.

Again, let's assume that we have a population of $p(t_0)=p_0$ in the beginning. This is called an initial condition. Then we assume that as time ticks on that the change in population per time unit is proportional to the population itself, so:
$$
\frac{d}{dt}p(t)=kp(t)
$$
, where $k$ is the proportionality constant and $k>0$ for population ..., $k<0$ for population ..., and $k=0$ for population .... This is the actual differential equation.

And we know that the solution to this for $k\neq 0$ is the exponential function:

$$
p(t)=p_0e^{kt}
$$
, where we've made sure to satisfy that $...$!

### General definition of a first-order ODE

So, what is an ODE now? An ODE of the first order is a function of the form:

$$\frac{df(t)}{dt} = f'(t) = g(f(t))$$

So, actually, in an ODE we know that the change of a function is equal to some other (complex) function of that function $g(f(t))$. 

In addition to this implicit definition of $f(t)$, we also need an initial condition, say:

$$f(t_0)=f_0$$

This condition tells us what the solution of the system looks like when we start the clock. 

<div class='alert alert-warning'>
<p>Why first-order ODE? Because the ... derivative is the first derivative. So, higher-order ODEs simply have additional derivatives of higher order.
</div> 

<div class='alert alert-warning'>
<p>Why ordinary (**O**DE)? Because the function f(t) that we are looking for only depends on ..., and on ... variable. 
</div>

In general, as we will see, solving such equations analytically is very, very hard, so in the following, we will work with a few numerical methods to try to solve and characterize ODEs. 

## Numercial methods - Euler's method
Since we do not want to solve ODEs explicitly, we first need an algorithm that can tackle this task. 

By far the most simple method is **Euler's method**.

For this method, we start by remembering the definition of the derivative

$$f'(t_1) = \lim_{dt\to 0} \frac{f(t_1+dt)-f(t_1)}{dt}$$

So, approximately (without the limit), the derivative of a function at a point $t_1$ is:

$$f'(t_1) \approx \frac{f(t_1+dt)-f(t_1)}{dt}$$

In general, we remember from above that a first order ODE is:

$$\frac{df(t)}{dt} = f'(t) = g(f(t))$$

Combining the two equations, we get for a specific point $t_1$:

$$g(f(t_1)) = \frac{f(t_1+dt)-f(t_1)}{dt}$$


And, solving for $f(t_1+dt)$, we get

$$f(t_1+dt) = g(f(t_1))*dt + f(t_1) + \text{Error}$$

Ignoring the error term, which comes from the derivative approximation, this is Euler's algorithm! 

It is an ... algorithm that allows us to move the solution to the equation a little step $dt$ forward, given that we can evaluate $g(f(t_1))$ and $f(t_1)$.

Now, of course there's a problem: we do know $g(f(t_1))$, since that is given to us as the RHS of the ODE definition, but we do **NOT** know $f(t_1)$ in general. 

And this is where the **...** comes in. In order to solve the equation, we need to start at a certain point, which is determined by the ... (now we know why it's so important to specify those conditions!).

So, since we know that ..., we will start our iterative algorithm from there.  


## Exponential function

Let's do a simple example first. 

We would like to integrate the standard equation that gives us the exponential function.

So, we assume:

$$\frac{df}{dt} = f(t)$$

That is, the derivative of the function $f(t)$ is the function itself. 

In order to start some sort of numerical integration (and to resolve the unknown constant uniquely), we need of course to know some starting point. Let's say, we know that:

$$f(0)=a$$

, where we will set $a$ to several values in the following examples.

In the following code, let's try to implement Euler's method based on the equations given above.

```matlab
% define the equation properly
% parameter a - this is simply a constant
a = 1;

% now define the actual ODE with df/dt = g = ?
% this needs to be a function that receives a f-value
% and returns g, of course
g = @(f) ...;

% step size for Euler integration
dt = 0.0003;

% total number of integration steps
totNumSteps = 10000;

% here let's define two arrays: one called ft 
% and another called t. ft contains the solution
% of the ODE and t contains the t-values at which
% we evaluate the solution
% 1) think about how many points we need in each array
% 2) make two arrays ft and t with zeros

ft=zeros(...,1);
t=zeros(...,1);

% since Euler's method is iterative, we need starting
% values for both ft and t
% think about what these are and put them in:
 
ft(1)=...;
t(1)=...;

% good, now here we actually do the numerical integration
% so, we need a for-loop that goes through all the steps:

for i =...:...
    
    % in each step of the loop, we now integrate the
    % ODE according to Euler's method. 
    % For this, remember the definition of the derivative
    %      f'(t1) = (f(t1+dt)-f(t1))/dt
    % we do know:
    %      f(t1)     that's our starting point ft(1)
    %      dt        that's our step size
    %      f'(t1)    that's g, our function that we defined
    % we WANT to know:
    %      f(t1+dt)  that's our next point ft(t(2))
    % armed with this, you should be able to solve and
    % update the next values for ft and t
    
    % here goes the update for the next value of the function
    ft(i)=...;
    % here goes the update for the next time value
    t(i)=...;
end    

% now let's plot this and compare to the true solution
plot(t,ft,'b-'); hold on
plot(t,a*exp(t),'r--')
title(sprintf('SSE = %f',sum((a*exp(t)-ft).^2)/totNumSteps))
```

When "integrating" the ODE with parameters $a=1$, $dt = 0.0003$, and $\text{totNumSteps}=10000$, the curve looks like a nice exponential and the SSE per point is very, very low.

When playing around with different parameters, however, we will see that Euler's method is only an approximation and that as we approach large values of $f(t)$, we will make larger and larger errors! 

### Using `ode45` from Matlab

Now, let's use the professional toolkit! Matlab of course has a function called `ode45` that solves first-order ODEs for you using much more sophisticated algorithms (such as Runge-Kutta methods, etc.). We set this up in the following way:

```matlab
% set up the function, so that Matlab's solver can use it
% note that the function accepts TWO arguments, t and y
% here, we need only one of them to complete the task!
df_dt = @(t, y) ...; 

% total time we want to get our data for
t = [0:3e-4:3];            

% initial function value
f0 = ...;                    

% let's use the standard, recommended method in Matlab, ode45
[ts,ys] = ode45(df_dt, t, f0);

% now let's plot this and compare to the true solution
plot(ts,ys,'b-'); hold on
plot(ts,ys,'r--')
title(sprintf('SSE = %f',sum((exp(ts)-ys).^2)/length(ts)))
```

## Extending the simple example

Now, let's upgrade our population function to something nicer. Let's say, we know that food is limited and that there is a maximum population that our environment can support.

So, the ODE of growth we had before:

$$\frac{df}{dt} = f(t)$$

does not hold anymore. We need to change it.

What we need is a function, whose change goes to 0 as soon, as the function value (i.e., the population) approaches a limit. We call this limit $K$ the carrying capacity of the environment. 

So, naively, we could say that 

$$\frac{df}{dt} = f(t)-K$$

But is this good? No, because initially, if our population is not $K$, the growth will be negative! We also need something where the initial growth (if $f(t)<<K$) is almost exponential. 

This means, we need two terms that fight against each other, which points to a multiplicative solution.

So, our total function becomes:

$$\frac{df}{dt} = ...$$

where $K$ is carrying capacity.

```matlab
%%file df_dt_carrying.m
% function that implements the input function for a limited
% capacity environment
% 
% y is the current function value 
% t is the current time point - for now we don't need this
% K is an additional parameter that indicates 
% the carrying capacity of our model

% Change the following code to return the correct function 
% value for the ODE we want to investigate
function output = df_dt_carrying(t,y,K) 
output = y.*(1-y/K);
```

```matlab
% total time we want to get our data for
t = linspace(0, 20,  100);            

% initial function value
f0 = 0.1;             

% carrying capacity
K = 1000;

% let's call ode45 again
% note that we have to pass "K" implicitly to ode45
% since this supports only TWO parameters!
[ts,fs] = ode45(@(t,y) df_dt_carrying(t,y,K), t, f0);

plot(ts,fs); hold on
plot(ts,0.1*exp(ts),'r--')
plot([0,20],[1000,1000],'k--')
xlabel('time'); ylabel('population')
ylim([0 1100])
grid on
```

And we can see how similar to the exponential the initial growth is, but that it rapidly slows down as the capacity limit is reached.

### The logistic function

For those scratching their head and wondering - the differential equation above for $K=1$ looks like:

$$f' = f(1-f)$$

and we have seen this before! In fact, when we looked at neural networks, we saw that the derivative of the logistic activation function $f(t) = \frac{1}{1+e^{-t}}$ has a derivative of that form.

Hence, the logistic function is the solution to that equation for initial conditions $f_0$ and carrying capacity $K$:

$$f(t) = \frac{K}{1+\frac{K-f_0}{f_0}e^{-t}}$$

This is also called the logistic equation.

The solution to this equation actually can be derived by integration of parts analytically - let's take the case that $K=1$ for simplicity, so we have:

$$\begin{align}
\frac{df}{dt}&=f(1-f)\\
\int \frac{df}{f(1-f)} &= \int dt\\
\int \frac{1}{f}df + \int \frac{1}{1-f}df &= t\\
...\\
...\\
...\\
\rightarrow f &= \frac{1}{1+e^{-t}}\\
\end{align}
$$

## Rabbits and wolves
Next, let's take a look at the Lokta-Volterra model, also known as the predator-prey equations, which are a pair of first order, non-linear, differential equations frequently used to describe the dynamics of biological systems in which two species interact, one a predator and one its prey. 

They were proposed independently by Alfred J. Lotka in 1925 and Vito Volterra in 1926 and they go like so:

$$
\begin{align}
\frac{dR}{dt} &=  k*R &-   a*R*W\\
\frac{dW}{dt} &= -r*W &+   b*R*W\\
\end{align}
$$

with the following notations:

* $R$: number of preys (for example, rabbits)
* $W$: number of predators (for example, wolves)  
* $k, a, r, b$ as constant parameters defining the behavior of the population:
    * $k$ is the natural growing rate of rabbits, when there's no wolf
    * $a$ is the natural dying rate of rabbits, due to predation
    * $r$ is the natural dying rate of a wolf, when there's no rabbit
    * $b$ is the factor describing how many caught rabbits can support a new wolf

It is convenient to combine the two populations in one vector as follows:
$$\vec{X}=(R,W)$$



```matlab
%%file df_dt_LV.m
% This code will combine the two equations in one array:
% X=[R, W] to describe the state of both populations.

function output = df_dt_LV(t,X)
% Definition of parameters, these are selected to 
% provide meaningful numbers for rabbit and wolf
% "interactions"
k = 0.08;
a = 0.001;
r = 0.02;
b = 0.00002;

output =  [...];
```

### Population equilibrium  
Before using solving this system, we will have a closer look at the equilibrium. 

By definition, an equilibrium occurs when the growth rate $\frac{d\vec{X}}{dt}$ is equal to 0. This gives two fixed points:

```matlab
k = 0.08;
a = 0.001;
r = 0.02;
b = 0.00002;

% first fixed point (boring - everything is 0)
X_f1 = [0;0];
% second fixed point (less boring - when dR/dt and dW/dt =0)
X_f2 = [...;...];
fprintf('Second fixed point has %f rabbits and %f wolves\n',X_f2(1),X_f2(2))

% both points should of course have 0 growth rate,
% so let's check this explicitly:
all(df_dt_LV(0,X_f1) == 0) & all (df_dt_LV(0,X_f2) == 0)
```

Whereas the first point is rather boring, the second point is much more interesting. Apparently for a finite number of predators and prey, the system will forever stay at this population vector - the number of rabbits eaten is just enough to feed this amount of wolves, given their reproductive rate.

### Stability of the fixed points
Near these two points, the system can be linearized:
$$\frac{d\vec{X}}{dt} \approx \mathbf{A}*\vec{X}$$
where $\mathbf{A}$ is the so-called Jacobian matrix evaluated at the corresponding point - since we need derivatives in every direction $R,W$:

$$A=\left(
\begin{array}{cc}
\frac{dR/dt}{dR}&\frac{dR/dt}{dW}\\
\frac{dW/dt}{dR}&\frac{dW/dt}{dW}\\
\end{array}
\right)$$

This is an extension of the simple Taylor expansion for multivariate functions.

```matlab
k = 0.08;
a = 0.001;
r = 0.02;
b = 0.00002;

% let's define the Jacobian matrix
d2X_dt2 = @(X) [[...,...];
                [...,...]];  
                
                 
% So, near X_f0, which represents the extinction of 
% both species, we will have the following changes
% in population:
A_f1 = d2X_dt2(X_f1)

% Near X_f0, the number of rabbits increases 
% and the population of wolves decreases.
% The origin is therefore a saddle point!
 
% Near X_f1, we have:
A_f2 = d2X_dt2(X_f2) 

% It is interesting to look at the eigenvalues of 
% the Jacobian (which, as we will see are +/- sqrt(k*r)*j)
[eigvecs, eigvals] = eig(A_f2); 
disp(eigvals(1,1));disp(eigvals(2,2));

% They are imaginary numbers, so that the wolf and 
% rabbit populations are periodic and their approximate 
% period will be given by:
T_f2 = 2*pi/abs(eigvals(1))
```

### Integrating the ODE using `ode45` 
Now we will use Matlab's ode45 to integrate the ODEs. In addition, we will be using the solution to try and find the actual period.

```matlab
% total time we want to get our data for
t = linspace(0, 1500,  1000);            

% initial population of rabbits and wolves
X0 = [1000;40];                    

% let's make Matlab a bit more precise
opts = odeset('RelTol',1e-5,'AbsTol',1e-7);
[ts,Xs] = ode45(@df_dt_LV, t, X0, opts);

% population solutions
rabbits=Xs(:,1);
wolves =Xs(:,2);

% find the peaks in the signal
[~,rpeaks] = findpeaks(rabbits);
[~,wpeaks] = findpeaks(wolves);


% since we have a periodic signal, the period is 
% (hopefully) the average distance between peaks
rperiod=...;
wperiod=...;


yyaxis left
plot(t, rabbits, 'displayname','rabbits'); 
xlabel('time'); ylabel('rabbit population')
yyaxis right
plot(t, wolves, 'displayname','wolves')
ylabel('wolf population')

title(sprintf('Rabbit and wolf populations\nRabbit period %.2f, Wolf period %.2f',rperiod,wperiod))

```

### Plotting direction fields and trajectories in the phase plane
We will plot some trajectories in a so-called "phase plane" for different starting points between X_f1 and X_f2. This will allow us to trace how the two population interact.

```matlab
k = 0.08;
a = 0.001;
r = 0.02;
b = 0.00002;

t = linspace(0, 1500,  1000);

% first fixed point (boring - everything is 0)
X_f1 = [0;0];
% second fixed point (less boring - when dR/dt and dW/dt =0)
X_f2 = [...;...];

% we wil evaluate points between the two
% equilibrium points X_f0 and X_f1
values  = linspace(0.3, 0.9, 5);

% get colors for each trajectory
cm = colormap('autumn');
vcolors = cm(ceil(linspace(1,length(cm)-10,length(values))),:);  
opts = odeset('RelTol',1e-5,'AbsTol',1e-7);
% plot trajectories
for i=1:length(values) 
    % starting point
    X0 = values(i) * X_f2;
    % integrate the ODE for the times and starting points
    [ts,Xs] = ode45(@df_dt_LV, t, X0, opts);
    % plot the trajectory with varying linewidth and color
    plot( Xs(:,1), Xs(:,2), 'LineWidth',2,'Color',vcolors(i,:), 'displayname',sprintf('X0=%.2f, %.2f',X0(1), X0(2))); hold on
    plot( X0(1), X0(2), 'xk','HandleVisibility','off')
end    
legend

% now define a few rs, ws
rxs = linspace(10,6000,30);
wys = linspace(2,220,30);

% for each of these population values...
for i=1:length(rxs)
    for j=1:length(wys)
        % ... determine the direction of evolution
        % at that point, which is of course given
        % by the ODE itself (derivatives are useful!)
        tmp = df_dt_LV(0,...);
        rxsm(i,j)=rxs(i);
        wysm(i,j)=wys(j);
        % normalize the gradient
        g1(i,j)=tmp(1)/norm(tmp);
        g2(i,j)=tmp(2)/norm(tmp);
    end
end
% plot all gradients as quiver arrows into final plot
quiver(rxsm,wysm,g1,g2,0.4,'displayName','gradient');
grid on
xlabel('Rabbits'); ylabel('Wolves');
xlim([0 6000]); ylim([0 220])
```

We can see from the phase space, the different evolution of the trajectories. We can see that they all are periodic, that they all are revolving around a fixed point, which is shown in the middle at the position calculated above. We can also see where the population changes are very fast and where they are slower. 

## The real pendulum

For the real pendulum, we have a force that will pull it back due to gravity. The component of gravity that does this depends on the angle of the pendulum, of course. 

The component of the gravitational force, which does this, becomes:

$$
F_g = -mg\sin{\alpha}
$$

, where F has a negative sign as it pulls "back".

Now this force will accelerate the point mass on the bottom with $F = ma$. There's a slight hiccup here, since we are talking about the **angular acceleration**. The position of the point on the circular arc it makes is given by its arc length $s$ defined as:

$$
s = ...
$$

, where $l$ is the length of the pendulum (longer pendulums will trace out a longer path on a circle for the same angle).

So, with this, angular speed is the first derivative:

$$
v = \frac{ds}{dt} = ...
$$

and angular acceleration is the second derivative:

$$
a = \frac{dv}{dt} = \frac{d^2s}{dt^2}=...
$$

So in total we get:

$$
m... = -mg\sin{\alpha}
$$

from which the mass cancels and therefore

$$
... = 0
$$

So, we are looking for a function whose second derivative is the **sin** of the function. This is ugly. 

In introductory physics, you will usually use the Taylor-expansion of the sin-function to set $\sin{\alpha} \approx \alpha$, from which you will get the canonical form of a harmonic oscillator with trigonometric functions as its solutions (the solution of $\frac{d^2\alpha}{dt^2} + \frac{g}{l}\alpha = 0$ is obviously sin/cos functions!). You can only make this assumption for small angles though, so if the pendulum extends further than, say, 6 degrees this is clearly wrong!

There are no known solutions to the general pendulum equation, so we resort to numeric treatment.

Now there's another issue. In general, the ODE-solvers can only handle **first-order** ODEs!

How can we do this?? Well, we can convert our problem that has one equation to **two** equations!

So, instead of $\frac{d^2\alpha}{dt^2} +\frac{g}{l}\sin{\alpha} = 0$, we write our ODE as a system of coupled, first-order ODEs that use the derivatives of the angle and angular speed like so:

$$
...\\
...
$$

```matlab
%plot inline -w 600 -h 2000 -r 150

% gravity and length of pendulum (units are meters,seconds)
g = 9.81;
l = .5;

% initial state of the system with angle, speed
X0 = [[.01*pi;0] [.01*pi;1] [.5*pi;0] [.8*pi;0]];

% time for solving
t = linspace(0,10,1000);

% define the ODE for the pendulum
pODE = @(t,X) [...;...];
figure;
for x = 1:4

    % solve
    opts = odeset('RelTol',1e-5,'AbsTol',1e-7);
    [ts,Xs]=ode45(pODE,t,X0(:,x),opts);

    % get angles and angular speed and plot
    a = Xs(:,1); v = Xs(:,2);
    subplot(5,1,x);
    yyaxis left
    plot(ts,a);
    ylabel('angle')
    yyaxis right
    hold on;plot(ts,v);
    ylabel('speed')
    legend
    title(sprintf('a0=%.2f, v0=%.2f',X0(1,x),X0(2,x)))
    subplot(5,1,5)
    hold on
    plot(a,v);
    grid on; xlabel('angle'); ylabel('speed')
    title('Phase space of pendulum')
end
```

And we can see how for small angles, this really looks like a sin-function, but how for large angles this breaks down.
